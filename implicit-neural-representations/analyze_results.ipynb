```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
filename = '../experiments/sr1_exp_0.csv' 

pts = [55, 7, 8, 9]
```

# Calculate the contrast metric

```python
# Arbitrary change to add to initial commit - TEST BRANCH develop_airosado1
df = pd.read_csv(filename)
df_all = pd.read_csv('../experiments/sr1_exp_5.csv')
df = df.loc[(df.metric == "C") & (df.epoch == 2950)]
newdf = df.loc[(df.patient == 8)]
```

```python
df2 = newdf.drop(['epoch', 'seed'], axis = 1).groupby(['patient', 'image','direction']).mean()#.plot(kind = "bar")
```

```python
df3 = pd.concat([df, df_all], axis=0)
df3 = df3.loc[(df3.metric == "C") & (df3.epoch == 2950) & (df3.image != 'reconst')]
```

```python
df3
```

```python
for pt in pts:
    newdf = df3.loc[(df3.patient == pt)]
    sns.barplot(data=newdf, x="direction", y="performance", hue="image", palette="mako")
    plt.title("patient {}".format(pt))
    plt.legend(bbox_to_anchor=(1.02,1), loc='upper left', borderaxespad=0)
    plt.show()
```

# Calculate CNR

```python
df = pd.read_csv(filename)
df = df.loc[(df.metric == "CNR") & (df.epoch == 2950)]
for pt in pts:
    newdf = df.loc[(df.patient == pt)]
    sns.barplot(data=newdf, x="direction", y="performance", hue="image", palette="mako")
    plt.title("patient {}".format(pt))
    plt.legend(bbox_to_anchor=(1.02,1), loc='upper left', borderaxespad=0)
    plt.show()
```
