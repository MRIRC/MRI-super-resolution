```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
filename = '../experiments/sr1_exp_4.csv' 

pts = [55, 7, 8, 9]
```

# Calculate the contrast metric

```python
df = pd.read_csv(filename)
df = df.loc[(df.metric == "C") & (df.epoch == 2950) & (df.image != 'reconst') ]
```

```python
for pt in pts:
    newdf = df.loc[(df.patient == pt)]
    sns.barplot(data=newdf, x="direction", y="performance", hue="image", palette="rocket")
    plt.title("patient {}".format(pt))
    plt.show()
```

# Calculate CNR

```python
df = pd.read_csv(filename)
df = df.loc[(df.metric == "CNR") & (df.epoch == 2950) & (df.image != 'reconst') ]
for pt in pts:
    newdf = df.loc[(df.patient == pt)]
    sns.barplot(data=newdf, x="direction", y="performance", hue="image", palette="rocket")
    plt.title("patient {}".format(pt))
    plt.show()
```
